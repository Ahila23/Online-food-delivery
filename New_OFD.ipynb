**Handling missing values using mean, median, or** **mode**

```python
#Column 1:Order_ID,Column 2:Customer_ID,Column 7:Restaurant_ID,Column 8:Restaurant_Name,Column 18:Order_Status,Column 20:Delivery_Partner_ID,
#Column 22:Restaurant_Rating,Column 23:Order_Day,Column 25:Profit_Margin are cleaned data


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Step 1: Replace 0 with NaN
df['Customer_Age'] = df['Customer_Age'].replace(0, np.nan)

# Step 2: Fill using gender median
df['Customer_Age'] = df['Customer_Age'].fillna(
    df.groupby('Customer_Gender')['Customer_Age'].transform('median')
)

# Step 3: Fill remaining using overall median
df['Customer_Age'] = df['Customer_Age'].fillna(df['Customer_Age'].median())

# Step 4: Convert to int
df['Customer_Age'] = df['Customer_Age'].astype(int)

print("Remaining Missing:", df['Customer_Age'].isna().sum())
plt.hist(df['Customer_Age'])
print(df['Customer_Age'])
```

```python
# Convert to integer where appropriate (after handling NaNs) fir Customer_age
int_column=['Customer_Age']
print(int_column)
for col in int_column:
  df[col]=df[col].fillna(0).astype(int)
df
```

```python
# Convert categorical/object columns to 'category' for efficiency
cat_column=['Customer_Gender','City','Area','Cuisine_Type','Payment_Mode','Peak_Hour']
for col in cat_column:
  df[col]=df[col].astype('category')
  print(df.dtypes)
```

```python
# Filling the Null values for Customer_Gender Column with the most frequent value (mode)
df['Customer_Gender'] = df['Customer_Gender'].fillna(df['Customer_Gender'].mode()[0])
print(df['Customer_Gender'])
```

```python
#Fill city values with mode

#Step1:Clean the City Column Properly
df['City'] = df['City'].astype(str).str.strip().str.title()

#Step2:Convert Fake Missing Values to Real NaN
import numpy as np

df['City'] = df['City'].replace(['Nan', 'None', 'Null', ''], np.nan)

#Step3:Now Apply Mode
df['City'] = df['City'].fillna(df['City'].mode().iloc[0])

#Step4:Check results
print(df['City'].value_counts())

print(df['City'])
```

```python
#Fill Area values with mode
df['Area']= df['Area'].fillna(df['Area'].mode()[0])
print(df['Area'])
```

```python
#Fill Cuisine_Type with mode
df['Cuisine_Type']=df['Cuisine_Type'].fillna(df['Cuisine_Type'].mode()[0])
print(df['Cuisine_Type'])
```

```python
#convert Order_Date to datetime
df['Order_Date']=pd.to_datetime(df['Order_Date'],errors='coerce')
print(df['Order_Date'])

#Fill with Most Frequent Date for missing dates
df['Order_Date'] = df['Order_Date'].fillna(df['Order_Date'].mode()[0])

#To check missing values
print(df['Order_Date'].isna().sum())
```

```python
#handling missing values for Delivery_Rating by using fill

# Convert rating column to numeric
df['Delivery_Rating'] = pd.to_numeric(df['Delivery_Rating'], errors='coerce')

# Check missing values
print("Missing before cleaning:", df['Delivery_Rating'].isnull().sum())

# Fill missing values with median rating
df['Delivery_Rating'] = df['Delivery_Rating'].fillna(df['Delivery_Rating'].median())

# Check again
print("Missing after cleaning:", df['Delivery_Rating'].isnull().sum())

# Convert to integer (optional)
df['Delivery_Rating'] = df['Delivery_Rating'].astype(int)

print(df['Delivery_Rating'].head())
```

```python
#Fill null values for Delivery_Time_Min by using imputation Techniques
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Ensure numeric
df['Delivery_Time_Min'] = pd.to_numeric(df['Delivery_Time_Min'], errors='coerce')

# Impute
imputer = SimpleImputer(strategy='mean')
df['Delivery_Time_Min'] = imputer.fit_transform(df[['Delivery_Time_Min']]).ravel()

# Scale
scaler = StandardScaler()
df['Delivery_Time_Min'] = scaler.fit_transform(df[['Delivery_Time_Min']]).ravel()

print(df['Delivery_Time_Min'])
```

```python
# Filling missing values for Distance_km
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Ensure numeric
df['Distance_km'] = pd.to_numeric(df['Distance_km'], errors='coerce')

# Median imputation
imputer = SimpleImputer(strategy='median')
df['Distance_km'] = imputer.fit_transform(df[['Distance_km']]).ravel()

# Scale
scaler = StandardScaler()
df['Distance_km'] = scaler.fit_transform(df[['Distance_km']]).ravel()


print("Missing values after imputation:",
df['Distance_km'].isna().sum())

print(df['Distance_km'])
df.head()
```

```python
#Filling null values for Order_Time
import pandas as pd
import numpy as np

# Step 1: Convert to datetime
df['Order_Time'] = pd.to_datetime(df['Order_Time'], errors='coerce')

# Step 2: Fill missing using most frequent time per Restaurant
df['Order_Time'] = df['Order_Time'].fillna(
    df.groupby('Restaurant_ID')['Order_Time']
      .transform(lambda x: x.mode()[0] if not x.mode().empty else pd.NaT)
)

# Step 3: Fill remaining using overall most frequent time
df['Order_Time'] = df['Order_Time'].fillna(df['Order_Time'].mode()[0])
print(df['Order_Time'])

# Step 4: Check
print("Remaining Missing:", df['Order_Time'].isna().sum())
```

```python
#Filling Nan for Peak_Hour


import numpy as np

# Step 1: Convert empty to NaN (if needed)
df['Peak_Hour'] = df['Peak_Hour'].replace('', np.nan)

# Step 2: Fill missing using most frequent value per Order_Day
df['Peak_Hour'] = df['Peak_Hour'].fillna(
    df.groupby('Order_Day')['Peak_Hour']
      .transform(lambda x: x.mode()[0] if not x.mode().empty else False)
)

# Step 3: If still any missing, fill with False
df['Peak_Hour'] = df['Peak_Hour'].fillna(False)

# Step 4: Convert to boolean
df['Peak_Hour'] = df['Peak_Hour'].astype(bool)
print(df['Peak_Hour'])


print("Remaining Missing:", df['Peak_Hour'].isna().sum())

```

**Removing or capping outliers (delivery time, order value)**

```python

df_temp=df.copy()
# Select only numeric columns
numeric_cols = df_temp.select_dtypes(include=['int64', 'float64']).columns

# Loop through each numeric column and detect outliers using IQR
for col in numeric_cols:
    Q1 = df_temp[col].quantile(0.25)   # First quartile (25%)
    Q3 = df_temp[col].quantile(0.75)   # Third quartile (75%)
    IQR = Q3 - Q1                      # Interquartile Range

    lower_bound = Q1 - 1.5 * IQR       # Lower limit for outliers
    upper_bound = Q3 + 1.5 * IQR       # Upper limit for outliers

    outliers = df_temp[(df_temp[col] < lower_bound) | (df_temp[col] > upper_bound)]

    print(f"📌 Column: {col} — Found {len(outliers)} outliers")
```

```python
# ==============================
# LOGICAL CONSISTENCY CLEANING
# ==============================

# 1️⃣ Standardize Order_Status text
df['Order_Status'] = df['Order_Status'].astype(str).str.strip().str.title()

# 2️⃣ Clip ratings between 1 and 5
df['Restaurant_Rating'] = df['Restaurant_Rating'].clip(1, 5)
df['Delivery_Rating'] = df['Delivery_Rating'].clip(1, 5)

# 3️⃣ Remove ratings & delivery time for Cancelled orders
import numpy as np

df.loc[df['Order_Status'] == 'Cancelled',
       ['Delivery_Rating', 'Restaurant_Rating', 'Delivery_Time_Min']] = np.nan

# 4️⃣ Ensure Cancelled orders have Cancellation Reason
df.loc[(df['Order_Status'] == 'Cancelled') &
       (df['Cancellation_Reason'].isna()),
       'Cancellation_Reason'] = 'Not Specified'

# 5️⃣ Ensure Completed orders do NOT have Cancellation Reason
df.loc[df['Order_Status'] == 'Completed',
       'Cancellation_Reason'] = None

# 6️⃣ Fix negative discount values
df['Discount_Applied'] = df['Discount_Applied'].clip(lower=0)

# 7️⃣ Ensure Final_Amount calculation is correct
df['Final_Amount'] = df['Order_Value'] - df['Discount_Applied']

# 8️⃣ Remove negative profit
df['Profit_Margin'] = df['Profit_Margin'].clip(lower=0)

# 9️⃣ No profit for Cancelled orders
df.loc[df['Order_Status'] == 'Cancelled',
       'Profit_Margin'] = 0

# 🔟 Final Validation Summary
print("✅ Data Cleaning Completed")
print("\nRemaining Issues Check:")

print("Cancelled orders with ratings:",
      df[(df['Order_Status']=='Cancelled') &
         (df['Delivery_Rating'].notna())].shape[0])

print("Completed orders with cancellation reason:",
      df[(df['Order_Status']=='Completed') &
         (df['Cancellation_Reason'].notna())].shape[0])

print("Negative Profit rows:",
      df[df['Profit_Margin'] < 0].shape[0])
```

```python
# EDA analysis for Weekend vs weekday demand
# ==========================================
# FULL EDA: WEEKEND VS WEEKDAY DEMAND
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# 1️⃣ Prepare Data
# -------------------------------

df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Order_Day'] = df['Order_Date'].dt.day_name()

df['Day_Type'] = df['Order_Day'].apply(
    lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday'
)

# -------------------------------
# 2️⃣ Order Volume Analysis
# -------------------------------

order_volume = df.groupby('Day_Type').size()

plt.figure()
plt.plot(order_volume.index, order_volume.values, marker='o')
plt.title("Order Volume: Weekend vs Weekday")
plt.xlabel("Day Type")
plt.ylabel("Total Orders")
plt.show()

# -------------------------------
# 3️⃣ Revenue Analysis
# -------------------------------

revenue = df.groupby('Day_Type')['Final_Amount'].sum()

plt.figure()
plt.plot(revenue.index, revenue.values, marker='o')
plt.title("Revenue Comparison")
plt.xlabel("Day Type")
plt.ylabel("Total Revenue")
plt.show()

# -------------------------------
# 4️⃣ Average Order Value
# -------------------------------

avg_order_value = df.groupby('Day_Type')['Order_Value'].mean()

plt.figure()
plt.plot(avg_order_value.index, avg_order_value.values, marker='o')
plt.title("Average Order Value")
plt.xlabel("Day Type")
plt.ylabel("Avg Order Value")
plt.show()

# -------------------------------
# 5️⃣ Average Delivery Time
# -------------------------------

avg_delivery = df.groupby('Day_Type')['Delivery_Time_Min'].mean()

plt.figure()
plt.plot(avg_delivery.index, avg_delivery.values, marker='o')
plt.title("Average Delivery Time")
plt.xlabel("Day Type")
plt.ylabel("Avg Delivery Time (Min)")
plt.show()

# -------------------------------
# 6️⃣ Cancellation Rate
# -------------------------------

cancel_rate = df.groupby('Day_Type')['Order_Status'] \
                .value_counts(normalize=True) \
                .unstack()

print("\nCancellation Rate:")
print(cancel_rate)

# -------------------------------
# 7️⃣ Profit Analysis
# -------------------------------

profit = df.groupby('Day_Type')['Profit_Margin'].sum()

plt.figure()
plt.plot(profit.index, profit.values, marker='o')
plt.title("Total Profit Comparison")
plt.xlabel("Day Type")
plt.ylabel("Total Profit")
plt.show()

# -------------------------------
# 8️⃣ Final Summary Table
# -------------------------------

summary = df.groupby('Day_Type').agg({
    'Order_ID': 'count',
    'Final_Amount': 'sum',
    'Order_Value': 'mean',
    'Delivery_Time_Min': 'mean',
    'Profit_Margin': 'sum'
}).rename(columns={
    'Order_ID': 'Total_Orders',
    'Final_Amount': 'Total_Revenue',
    'Order_Value': 'Avg_Order_Value',
    'Delivery_Time_Min': 'Avg_Delivery_Time',
    'Profit_Margin': 'Total_Profit'
})

print("\nFinal Summary Table:")
print(summary)
```

# **Feature Engineering**

```python
#order day type weekday/weekend

# ==========================================
# DERIVED COLUMN: WEEKDAY / WEEKEND
# ==========================================

import pandas as pd

# 1️⃣ Convert Order_Date to datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

# 2️⃣ Extract Day Name
df['Order_Day'] = df['Order_Date'].dt.day_name()

# 3️⃣ Create Day_Type column
df['Day_Type'] = df['Order_Day'].apply(
    lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday'
)

# 4️⃣ Check result
print(df[['Order_Date', 'Order_Day', 'Day_Type']].head())

#Box plot for demand comparsion
import matplotlib.pyplot as plt

demand = df['Day_Type'].value_counts()

plt.bar(demand.index, demand.values)
plt.title("Weekend vs Weekday Demand")
plt.xlabel("Day Type")
plt.ylabel("Number of Orders")
plt.show()


#to compare order value
import seaborn as sns

sns.boxplot(x='Day_Type', y='Order_Value', data=df)
plt.title("Order Value Distribution: Weekend vs Weekday")
plt.show()
```

```python
#======Peak hour indicator========


# Step 1:convert Order_Time todateti

import pandas as pd

# Convert to datetime (already converted )
df['Order_Time'] = pd.to_datetime(df['Order_Time'], errors='coerce')

# Extract hour
df['Order_Hour'] = df['Order_Time'].dt.hour


# Step 2: Create peak hour indicator
df['Peak_Indicator'] = df['Order_Hour'].apply(lambda x: 'Lunch Peak' if 12 <= x <= 15 else ('Dinner Peak' if 19 <= x <= 22
          else 'Non-Peak')
)

#Step 3: Checking results
print(df[['Order_Time', 'Order_Hour', 'Peak_Indicator']].head())

# Step 4:📊 EDA: Orders During Peak vs Non-Peak

import matplotlib.pyplot as plt

peak_counts = df['Peak_Indicator'].value_counts()

# Generate dynamic colors
colors = plt.cm.Set3(range(len(peak_counts)))

plt.pie(peak_counts.values,
        labels=peak_counts.index,
        autopct='%1.1f%%',
        colors=colors,
        shadow=True,
        startangle=90)

plt.title("Peak vs Non-Peak Order Distribution")
plt.axis('equal')
plt.show()




```
